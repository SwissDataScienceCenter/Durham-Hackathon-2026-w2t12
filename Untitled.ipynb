```python
import os, re, glob, json, pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

# ---- load data (self-contained) ----
candidates = ["analysis-of-corporate-reports-dataset", "../analysis-of-corporate-reports-dataset",
              os.path.expanduser("~/analysis-of-corporate-reports-dataset"), "/home/jovyan/analysis-of-corporate-reports-dataset"]
folder = next((p for p in candidates if os.path.isdir(p)), None)
files = sorted(glob.glob(f"{folder}/*.json"))
NARRATIVE, MAX_CHARS = ["item_1", "item_1A", "item_7"], 60000
def join_items(d, keys):
    t = "\n".join(d.get(k) for k in keys if isinstance(d.get(k), str) and d.get(k).strip())
    return t[:MAX_CHARS]
rows = []
for p in files:
    try:
        d = json.load(open(p))
        m = re.search(r"(\d{4})", d.get("period_of_report") or "")
        rows.append({"year": int(m.group(1)) if m else None, "text": join_items(d, NARRATIVE)})
    except Exception:
        pass
df = pd.DataFrame(rows).dropna(subset=["year"]); df["year"] = df["year"].astype(int)

# ---- your keyword lists ----
E_KEYWORDS = ["environment","environmental","climate","carbon","emission","emissions","greenhouse gas","ghg","energy","renewable","waste","water","pollution","recycling","biodiversity","sustainability","net zero"]
S_KEYWORDS = ["employee","employees","workforce","labor","human capital","health","safety","training","diversity","inclusion","community","human rights","wellbeing","talent"]
G_KEYWORDS = ["governance","board","director","audit","ethics","compliance","anti-corruption","bribery","risk management","internal control","cybersecurity","data privacy","shareholder","transparency"]

# normalise (lowercase, turn hyphens into spaces so phrases match the text)
norm = lambda k: k.lower().replace("-", " ").strip()
E, S, G = [list(map(norm, x)) for x in (E_KEYWORDS, S_KEYWORDS, G_KEYWORDS)]
pillar_words = {"E": E, "S": S, "G": G}
vocab = list(dict.fromkeys(E + S + G))   # unique keyword list

# ---- TF-IDF limited to those keywords, by year ----
by_year = df.groupby("year")["text"].apply(" ".join)
vec = TfidfVectorizer(vocabulary=vocab, ngram_range=(1, 2), stop_words=None)
matrix = vec.fit_transform(by_year)
vmap = vec.vocabulary_

# ---- pillar score per year = sum of its keyword weights ----
recs = []
for i, y in enumerate(by_year.index):
    r = matrix[i].toarray().flatten()
    rec = {"year": y}
    for name, words in pillar_words.items():
        rec[name] = sum(r[vmap[w]] for w in set(words) if w in vmap)
    recs.append(rec)
scores = pd.DataFrame(recs)
print(scores)

# ---- chart ----
plt.figure(figsize=(9, 5))
for col, label in [("E", "Environmental"), ("S", "Social"), ("G", "Governance")]:
    plt.plot(scores["year"], scores[col], marker="o", linewidth=2.5, label=label)
plt.xlabel("Year"); plt.ylabel("Total TF-IDF weight"); plt.legend()
plt.title("ESG language by pillar in S&P annual reports, 2016 to 2025")
plt.tight_layout(); plt.savefig("esg_pillars.png", dpi=150); plt.show()
```

```python
import os, re, glob, json, pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt

# ---- load data ----
candidates = ["analysis-of-corporate-reports-dataset", "../analysis-of-corporate-reports-dataset",
              os.path.expanduser("~/analysis-of-corporate-reports-dataset"), "/home/jovyan/analysis-of-corporate-reports-dataset"]
folder = next((p for p in candidates if os.path.isdir(p)), None)
files = sorted(glob.glob(f"{folder}/*.json"))
NARRATIVE, MAX_CHARS = ["item_1", "item_1A", "item_7"], 60000
def join_items(d, keys):
    t = "\n".join(d.get(k) for k in keys if isinstance(d.get(k), str) and d.get(k).strip())
    return t[:MAX_CHARS]
rows = []
for p in files:
    try:
        d = json.load(open(p))
        rows.append({"text": join_items(d, NARRATIVE)})
    except Exception:
        pass
df = pd.DataFrame(rows)

# ---- ESG keyword lists ----
E_KEYWORDS = ["environment","environmental","climate","carbon","emission","emissions","greenhouse gas","ghg","energy","renewable","waste","water","pollution","recycling","biodiversity","sustainability","net zero"]
S_KEYWORDS = ["employee","employees","workforce","labor","human capital","health","safety","training","diversity","inclusion","community","human rights","wellbeing","talent"]
G_KEYWORDS = ["governance","board","director","audit","ethics","compliance","anti-corruption","bribery","risk management","internal control","cybersecurity","data privacy","shareholder","transparency"]
norm = lambda k: k.lower().replace("-", " ").strip()
vocab = list(dict.fromkeys([norm(k) for k in E_KEYWORDS + S_KEYWORDS + G_KEYWORDS]))

# ---- count keyword usage across all filings ----
cv = CountVectorizer(vocabulary=vocab, ngram_range=(1, 2), stop_words=None)
counts = cv.fit_transform(df["text"])
totals = counts.sum(axis=0).A1
terms = cv.get_feature_names_out()

top = totals.argsort()[-10:][::-1]
words = [terms[i] for i in top]
freqs = [totals[i] for i in top]

# ---- bar chart ----
plt.figure(figsize=(9, 5))
plt.bar(words, freqs)
plt.xticks(rotation=45, ha="right")
plt.ylabel("Total count"); plt.title("10 most used ESG keywords across all filings")
plt.tight_layout(); plt.savefig("top10_esg_words.png", dpi=150); plt.show()
```
