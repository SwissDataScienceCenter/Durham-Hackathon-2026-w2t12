```python
import os, re, glob, json, pandas as pd

# find the data folder
candidates = [
    "analysis-of-corporate-reports-dataset",
    "../analysis-of-corporate-reports-dataset",
    os.path.expanduser("~/analysis-of-corporate-reports-dataset"),
    "/home/jovyan/analysis-of-corporate-reports-dataset",
]
folder = next((p for p in candidates if os.path.isdir(p)), None)
files = sorted(glob.glob(f"{folder}/*.json"))
print("folder:", folder, "|", len(files), "files")

# load only the narrative items
NARRATIVE = ["item_1", "item_1A", "item_7"]

def join_items(d, keys):
    return "\n".join(d.get(k) for k in keys if isinstance(d.get(k), str) and d.get(k).strip())

rows = []
for path in files:
    with open(path) as f:
        d = json.load(f)
    por = d.get("period_of_report") or ""
    m = re.search(r"(\d{4})", por)
    year = int(m.group(1)) if m else None
    rows.append({
        "company": d.get("company"),
        "sic": d.get("sic"),
        "year": year,
        "text": join_items(d, NARRATIVE),
    })

df = pd.DataFrame(rows).dropna(subset=["year"])
df["year"] = df["year"].astype(int)
print(df.shape)
print(df["company"].nunique(), "companies")
print(df.groupby("year").size())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

by_year = df.groupby("year")["text"].apply(" ".join)

vec = TfidfVectorizer(
    max_features=10000,
    stop_words="english",
    token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z]+\b",   # drops pure numbers
    min_df=2
)
matrix = vec.fit_transform(by_year)
terms = vec.get_feature_names_out()

for i, year in enumerate(by_year.index):
    row = matrix[i].toarray().flatten()
    top = row.argsort()[-15:][::-1]
    print(year, [terms[j] for j in top])
```

```python
import matplotlib.pyplot as plt

watch = ["climate", "sustainability", "emissions", "cybersecurity", "diversity", "pandemic"]
vocab = vec.vocabulary_

plt.figure(figsize=(9,5))
for w in watch:
    if w in vocab:
        weights = matrix[:, vocab[w]].toarray().flatten()
        plt.plot(by_year.index, weights, marker="o", label=w)
plt.xlabel("Year"); plt.ylabel("TF-IDF weight"); plt.legend()
plt.title("ESG and risk language in S&P annual reports, 2016 to 2025")
plt.tight_layout(); plt.savefig("esg_trend.png", dpi=150); plt.show()
```

```python
import os, re, glob, json, pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

# 1. find and load the data (capped per filing to keep memory safe)
candidates = [
    "analysis-of-corporate-reports-dataset",
    "../analysis-of-corporate-reports-dataset",
    os.path.expanduser("~/analysis-of-corporate-reports-dataset"),
    "/home/jovyan/analysis-of-corporate-reports-dataset",
]
folder = next((p for p in candidates if os.path.isdir(p)), None)
files = sorted(glob.glob(f"{folder}/*.json"))

NARRATIVE = ["item_1", "item_1A", "item_7"]
MAX_CHARS = 60000

def join_items(d, keys):
    txt = "\n".join(d.get(k) for k in keys if isinstance(d.get(k), str) and d.get(k).strip())
    return txt[:MAX_CHARS]

rows = []
for path in files:
    try:
        with open(path) as f:
            d = json.load(f)
        m = re.search(r"(\d{4})", d.get("period_of_report") or "")
        rows.append({"year": int(m.group(1)) if m else None,
                     "text": join_items(d, NARRATIVE)})
    except Exception:
        pass

df = pd.DataFrame(rows).dropna(subset=["year"])
df["year"] = df["year"].astype(int)
print("loaded", df.shape[0], "filings")

# 2. TF-IDF on single words, by year
by_year = df.groupby("year")["text"].apply(" ".join)
vec = TfidfVectorizer(max_features=10000, stop_words="english",
                      token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z]+\b", min_df=2)
matrix = vec.fit_transform(by_year)
vocab = vec.vocabulary_

# 3. ESG-only chart
esg_terms = {
    "Environmental": ["climate", "emissions", "carbon", "renewable", "sustainability"],
    "Social":        ["diversity", "inclusion", "workforce"],
    "Governance":    ["governance", "ethics"],
}
plt.figure(figsize=(10, 6))
for pillar, words in esg_terms.items():
    for w in words:
        if w in vocab:
            weights = matrix[:, vocab[w]].toarray().flatten()
            plt.plot(by_year.index, weights, marker="o", label=f"{w} ({pillar[0]})")
plt.xlabel("Year"); plt.ylabel("TF-IDF weight")
plt.title("ESG language in S&P annual reports, 2016 to 2025")
plt.legend(ncol=2); plt.tight_layout()
plt.savefig("esg_trend.png", dpi=150); plt.show()
```
